```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from matplotlib.colors import ListedColormap
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
#pridicted or dependent varible is the last column of the data set
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
```

```python
# Assuming the categorical feature is in the second column (index 1), #feature scaling, x is scaled, because of have non binary data which is converted to binary
# Create a ColumnTransformer to handle categorical and numerical features
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1]),  # Apply OneHotEncoder to the categorical feature at index 1
                                    ('scaler', StandardScaler(), [0, 2])],  # Apply StandardScaler to numerical features at indices 0 and 2
                       remainder='passthrough')  # Passthrough any remaining columns

# Fit and transform the training data
x_train = ct.fit_transform(x_train)

# Transform the testing data using the same transformer
x_test = ct.transform(x_test)

# Fit the Logistic Regression model
classifier = LogisticRegression(random_state=0)
classifier.fit(x_train, y_train)
```
